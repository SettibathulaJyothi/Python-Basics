**Input Function:**  
It is used to take input from the user.

```python
N1=input("Enter first number")
N2=input("Enter second number")
sum=N1+N2
print("sum= ",sum)
```

The reason for the above is that by default the input statement always takes input as string value.  
If a "+" is used with string,it concatenates the 2 string.In order to evaluate the expression mathematically,we must use int() with an input statement to convert it into integer.

```python
N1=int(input("Enter first number"))
N2=int(input("Enter second number"))
sum=N1+N2
print("sum= ",sum)
```
