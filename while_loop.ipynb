**The While Loop:**
It is used to repeat a set of instructions as long as the condition is True.

```python
count=1
while count<=5:
  print("Hello")
  count+=1
print("Program ends")
```

```python
i=1
while i<6:
  print(i)
  i+=1
```

```python
#program to reverse the digits of a number
num=int(input("Enter a number:"))
Digit=0
while num>0:
  R=num%10
  Digit=(Digit*10)+R
  num//=10
print(Digit)
```
