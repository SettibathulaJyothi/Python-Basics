#**Matplotlib:**  
Matplotlib was created by **John D.Hunter** in 2003.It is a free and open-source data visualization library used fro plotting graphs and visualization in python.Built on numpy array.  
It has various plotting modules,but the most commonly used module is pyplot,which provides an easy interface for creating 2D plots and visualizations.  Some types of graphs are
**Histogram, Pie Plot, Box Plot, Scatter Plot, Line Graph**
