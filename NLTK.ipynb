# **Some important commands of NLTK**  
**Tokenization**  
Tokenization is the process of converting large textual data into smaller parts called tokens.These  tokens help in nlp for finding patterns and are used for further processing through stemming and lemmatization.  
Tokenizing a sentence will split big sentences into smaller sentences.

```python
data="Hello Everybody!!!.Hope you re enjoying learning NLP.Wishing you a wonderful experince"
from nltk.tokenize import sent_tokenize
sent_token=sent_tokenize(data)
print(sent_token)
```

Tokenizing a word will split a sentence into words:

```python
word_token=nltk.word_tokenize(data)
print(word_token)
```

Stemming is the process of extracting base word from the given word:

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
ps.stem('studies')
```

```python
ps.stem('Learning')
```

# **Lemmatization**
It is done using WordNetLematizer package:

```python
from nltk.stem.wordnet import WordNetLemmatizer
Lem=WordNetLemmatizer()
Lem.lemmatize("studies")
```

# **Using NLTK Stopwords Corpus**  
Stop words are used to be removed to avoid noise in a sentence.

Stopwords are such as is,am,are,this,a,an,the,etc.,

```python
from nltk.corpus import stopwords
Eng_stopwords=set(stopwords.words('english'))
print(Eng_stopwords)
```
