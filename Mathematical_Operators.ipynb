**Using Mathematical Operators on Integer Array**  
Following table shows the use of mathematical operators on integer array:

**Operator Exponential** **


```python
import numpy as np
marks1=np.array([1,2,3,4])
marks2=np.array([2,1,2,1])
print(marks1**marks2)
print("marks1 to the power of 3 is :",marks1**3)
```

**Operator Addition +**

```python
import numpy as np
marks1=np.array([7,8,5,4])
marks2=np.array([5,6,4,7])
print(marks1+marks2)
print("2 marks extra:",marks1+2)
```

**Operator Subtraction -**

```python
import numpy as np
marks1=np.array([7,8,5,4])
marks2=np.array([5,6,4,7])
print(marks1-marks2)
print("2 marks deducted:",marks1-2)
```

**Operator Multiplication***

```python
import numpy as np
marks1=np.array([7,8,5,4])
marks2=np.array([5,6,4,7])
print(marks1*marks2)
print("2 marks multiplied:",marks1*2)
```

**Operator Divide /**

```python
import numpy as np
marks1=np.array([10,20,30,40])
marks2=np.array([2,4,3,5])
print(marks1/marks2)
print("2 marks multiplied:",marks1/2)
```

**Operator Floor Division //**

```python
import numpy as np
marks1=np.array([10,20,30,40])
marks2=np.array([2,4,3,5])
print(marks1//marks2)
print("2 marks divided using floor division:",marks1//2)
```

**Operator remainder of Division %**

```python
import numpy as np
marks1=np.array([10,20,30,40])
marks2=np.array([1,2,3,4])
print(marks1%marks2)
print("reminder of division:",marks1%3)
```
