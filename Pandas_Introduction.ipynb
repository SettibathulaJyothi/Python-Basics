**Pandas:**  
pandas is a an open source python library used for data manipulation and analysis.  
It provides strong features for working with three keys data structures:**Series**(1-Dimensional),**Dataframe**(2-Dimensional),and **Index**(used for label-based indexing).   
Pandas was created by **WesMcKinney** in 2008.The name"Pandas" is derived from Panel Data(a term in statistics for multidimensional data) and Python ,as it was designed fro data analysis in python.
