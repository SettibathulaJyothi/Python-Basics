**Accessing the Elements of the Series:**  
The elements can be accessed using either posional index or labelled index.  
For Example:


**To print labelled and positional index**

```python
import pandas as pd
Marks=pd.Series([96,92,91,93],index=["Shashi","Sonali","Neha","Rajeev"])
print("Using labelled index",Marks["Sonali"])
print("Using positional index",Marks.iloc[1])
```

Note:  
*use .iloc[pos] for positional indexing  
*use .loc[label] or [] for label-based indexing

**To print 2 elements:**

```python
print(Marks.iloc[[2,3]])
```

Note:  
*use .iloc[pos1,pos2] for positional indexing  
*use .loc[label1,label2] for label-based indexing

**To print 2 elements in another way:**

```python
print(Marks[["Sonali","Rajeev"]])
```

**To access elements using slicing:**

```python
print(Marks[1:3]) # excludes the value at indexing position 3
```

**To accesing elements using slicing in another way:**

```python
print(Marks["Sonali":"Rajeev"])
```

**To display the series in the reverse:**

```python
print(Marks[::-1])
```
