**DataFrames:**  
It is a 2D labellwd heterogeneous data datstructure that containns rows and columns.It has both a row and column index.It resembles the data arranged in a tabular from in a table of a database.

**Creating a dataframe**  
Different ways are shown below

**Creating an empty DataFrame**

```python
import pandas as pd
Emp=pd.DataFrame ()
print(Emp)
```

**Creating a DataFrame from Numpy Array**

```python
import numpy as np
import pandas as pd
a1=np.array([10,20,30])
a2=np.array([11,22,33])
f1=pd.DataFrame(a1)
f2=pd.DataFrame([a1,a2])
print(f1)
print(f2)
```
