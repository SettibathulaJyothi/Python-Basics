# **Plotting a Line chart**  
***Single line chart**

```python
!pip install matplotlib
```

```python
import matplotlib.pyplot as plt
city=['Delhi','Mumbai','Chennai','Kolkata']
temp=[35,21,40,22]
plt.xlabel('Metro cities')
plt.ylabel('Temp in Celsius')
plt.xlabel('Metro Cities')
plt.title('Temperature study of the day')
plt.plot(city,temp,marker='*',markersize=10,color='red',linewidth=2,linestyle='dashdot')
plt.show()
```

***Multiple line chart**

```python
import matplotlib.pyplot as plt
city=['Delhi','Mumbai','Chennai','Kolkata']
day1=[35,21,40,22]
day2=[30,25,36,20]
day3=[24,29,42,24]
plt.xlabel('Metro cities')
plt.ylabel('Temp in Celsius')
plt.title('Temperature study of the day')
plt.plot(city,day1)
plt.plot(city,day2)
plt.plot(city,day3)
plt.show()
```

# **Plotting a barchart**  
using bar()

```python
#data for the x-axis and y- axis
city=['Delhi','Mumbai','Chennai','Kolkata']
day1=[35,21,40,22]
#creating the bar chart
plt.bar(city,day1)
#adding labels and cities
plt.xlabel("Metro Cities")
plt.ylabel('Temp in Celsius')
plt.title('Temperature study of the day')
#displaying the chart
plt.show()
```

**Another example of bar chart**

```python
import matplotlib.pyplot as plt
marks=[23,45,34,41,13,49]
plt.bar(["Eng","Maths","Science","SSt","2nd laguage","Computer"],marks)
plt.title('Bar Chart')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.show()
```

**Using plot() for creating Single bar**

```python
import matplotlib.pyplot as plt
import pandas as pd
#create a dataframe
d1={'Cities':['Delhi','Mumbai','Chennai','Kolkata'],'Day1':[30,25,36,20]}
df=pd.DataFrame(d1)
#Creating the bar chart
df.plot(x='Cities',y='Day1',kind='bar')
#adding labels and title
plt.xlabel("Metro Cities")
plt.ylabel('Temp in Celsius')
plt.title('Temperature study of the day')
#displaying the chart
plt.show()
```

**Using plot() fro creating Multiple bars**

```python
import matplotlib.pyplot as plt
import pandas as pd
#create a dataframe
d1={'Cities':['Delhi','Mumbai','Chennai','Kolkata'],'Day1':[30,25,36,20],'Day2':[30,25,36,20],'Day3':[24,29,42,24]}
df=pd.DataFrame(d1)
df.plot(kind='bar')
#adding labels and title
plt.xlabel("Metro Cities")
plt.ylabel('Temp in Celsius')
plt.title('Temperature study of the day')
#displaying the chart
plt.show()
```

# **Plotting a histogram chart**

```python
import matplotlib.pyplot as plt
#number of schools in each city
schools=[10,25,29]
#creating histogram with bins
plt.hist(schools,bins=[10,15,20,25,30])
#Adding labels and title
plt.xlabel('Number of Schools')
plt.ylabel('Frequently')
plt.title('Distribution of schools in cities')
#customizing x-axis tick labels
plt.xticks([15,20,25,30])
#displaying the histogram
plt.show()
```

**Another example of histogram using hist() buit in function**

```python
from matplotlib import pyplot as pyplot
import numpy as np
fig,ax=plt.subplots(1,1)
marks=np.array([19,16,28,45,37,34,13,49,42,44])
ax.hist(marks,bins=[10,20,30,40,50])
ax.set_title("histogram of result")
ax.set_xlabel('marks')
ax.set_ylabel('no. of students')
plt.show()
```

# **Plotting a Pie Chart**

```python
import matplotlib.pyplot as plt
#Data for the pie chart
Temp=[20,30,15,10,25]
Cities=['Pune','Banglore','Jammu','Srinagar','Mumbai']
#Creating the pie chart
plt.pie(Temp,labels=Cities,autopct='%1.1f%%')
#Adding title
plt.title('Pie Chart')
#Dispalying the pie chart
plt.show()
```

# **Plotting a Scatter Chart**

```python
import matplotlib.pyplot as plt
marks1=[31,34,36,42,40,23,39]
marks2=[34,45,23,44,12,16,35]
plt.scatter(marks1,marks2,c='red')
plt.title('Scatter plot')
plt.xlabel('marks1')
plt.ylabel('marks2')
plt.show()
```

# **Plotting a Box Plot(Whisker Plot)**

```python
import matplotlib.pyplot as plt
eng=[82,76,24,40,67,62,78,67,72,82]
maths=[62,5,32,96,95,90,95,15,71,11]
science=[68,86,19,49,15,16,16,75,11]
sst=[98,10,87,29,72,16,23,72,75,30]
box_plot_data=[eng,maths,science,sst]
plt.xticks([1,2,3,4],["Eng","Maths","Science","sst"],rotation=10)
plt.boxplot(box_plot_data)
plt.title('Box Plot')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.show()
```
