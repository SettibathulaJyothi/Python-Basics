**Plotting a Line chart**  
***Single line chart**

```python
!pip install matplotlib
```

```python
import matplotlib.pyplot as plt
city=['Delhi','Mumbai','Chennai','Kolkata']
temp=[35,21,40,22]
plt.xlabel('Metro cities')
plt.ylabel('Temp in Celsius')
plt.xlabel('Metro Cities')
plt.title('Temperature study of the day')
plt.plot(city,temp,marker='*',markersize=10,color='red',linewidth=2,linestyle='dashdot')
plt.show()
```

***Multiple line chart**

```python
import matplotlib.pyplot as plt
city=['Delhi','Mumbai','Chennai','Kolkata']
day1=[35,21,40,22]
day2=[30,25,36,20]
day3=[24,29,42,24]
plt.xlabel('Metro cities')
plt.ylabel('Temp in Celsius')
plt.title('Temperature study of the day')
plt.plot(city,day1)
plt.plot(city,day2)
plt.plot(city,day3)
plt.show()
```

***Plotting a barchart**  
using bar()

```python
#data for the x-axis and y- axis
city=['Delhi','Mumbai','Chennai','Kolkata']
day1=[35,21,40,22]
#creating the bar chart
plt.bar(city,day1)
#adding labels and cities
plt.xlabel("Metro Cities")
plt.ylabel('Temp in Celsius')
plt.title('Temperature study of the day')
#displaying the chart
plt.show()
```

Another example of bar chart

```python
import matplotlib.pyplot as plt
marks=[23,45,34,41,13,49]
plt.bar(["Eng","Maths","Science","SSt","2nd laguage","Computer"],marks)
plt.title('Bar Chart')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.show()
```

**Using plot() for creating Single bar**

```python
import matplotlib.pyplot as plt
import pandas as pd
#create a dataframe
d1={'Cities':['Delhi','Mumbai','Chennai','Kolkata'],'Day1':[30,25,36,20]}
df=pd.DataFrame(d1)
#Creating the bar chart
df.plot(x='Cities',y='Day1',kind='bar')
#adding labels and title
plt.xlabel("Metro Cities")
plt.ylabel('Temp in Celsius')
plt.title('Temperature study of the day')
#displaying the chart
plt.show()
```
