**Type:** Type of an object i.e.array  

**Function:** type(ARR)

```python
import numpy as np
ARR=np.array((1,2,34))
print(type(ARR))
```

**Type:** Check the dimensions of an array

**Function:** ARR.ndim

```python
import numpy as np
ARR=np.array(([1,2,3,4],[3,4,5,6]))
print(ARR.ndim)
```

**Type:** Shape of an array i.e.length of the array dimensions

**Function:** ARR.shape

```python
import numpy as np
ARR=np.array([[1,2,3,4],[3,4,5,6]])
print(ARR.shape)
```

**Type:** Size of an array i.e.counts the number of elements.

**Function:** ARR.size

```python
import numpy as np
ARR=np.array([[1,2,3,4],[3,4,5,6]])
print(ARR.size)
```

**Type:** DataType of elements stored in the array

**Function:** ARR.dtype

```python
import numpy as np
ARR=np.array([[1,2,3,4],[3,4,5,6]])
print(ARR.dtype)
```
