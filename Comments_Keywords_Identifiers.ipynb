**Comments:**
Comments are used to increase the readability of the code.These are completely ignored by the python interpreter.They do not display or affect the output of the code.Its always a good practice to add coments in your code.Different types of comments are discussed below with examples.

**Single Line Comments:**
Starts with a hash symbol'#'followed by the text to be written as a comment that lasts till the end of the line.
