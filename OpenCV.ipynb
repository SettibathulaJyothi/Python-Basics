# **OpenCV**  
OpenCV stands for Open-Source Computer Vision Library,originally developed by Intel and oficially launched in 1999.  
It is an open source software library for computer vision and Machine Learning that helps a computer understand the content of digital images like photographs and videos.

**How to write a program in OpenCV**  
Following are the steps:  
**Import the required libraries**  
**Import matplotlib to display images**  
**Import numpy to store the pixel values in the NumPy array**

```python
#Display a 2D image using the imread() function of cv2.Thus,the complete code that appears
import cv2
from matplotlib import pyplot as plt
import numpy as np
img=cv2.imread('Flower1.jpeg')
plt.imshow(img)
plt.title('Flowers')
plt.axis('off')
plt.show()
```

```python
#Display a 2D image using the imread() function of cv2.Thus,the complete code that appears
import cv2
from matplotlib import pyplot as plt
import numpy as np
img=cv2.imread('Flower.jpeg')
plt.imshow(img)
plt.title('Flowers')
plt.axis('off')
plt.show()
```

Notice the colour of the image ,its because the opencv presents images in BGR(Blue,Green,Red) format ,but now we will convert it into RGB(red,green,blue)as follows  
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))

```python
#Display a 2D image using the imread() function of cv2.Thus,the complete code that appears
import cv2
from matplotlib import pyplot as plt
import numpy as np
img=cv2.imread('Flower.jpeg')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Flowers')
plt.axis('off')
plt.show()
```

```python
#Display a 2D image using the imread() function of cv2.Thus,the complete code that appears
import cv2
from matplotlib import pyplot as plt
import numpy as np
img=cv2.imread('Flower1.jpeg')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Flowers')
plt.axis('off')
plt.show()
```

To print the image in grayscale

```python
#Display a 2D image using the imread() function of cv2.Thus,the complete code that appears
import cv2
from matplotlib import pyplot as plt
import numpy as np
img=cv2.imread('Flower1.jpeg',0)
plt.imshow(img, cmap='gray')
plt.title('Flowers')
plt.axis('off')
plt.show()
```
