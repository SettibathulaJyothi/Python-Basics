**Attributes of the Series**  
Following shows some attributes of the series

**Attribute**:name  
**Description**:A name given to a series  given

```python
import pandas as pd
import numpy as np
data=np.array([10,30,50])
s1=pd.Series(data)
s1.name="Rollno"
print(s1)
```

**Attribute** :index.name  
**Description**:A name to an index is given

```python
s1.index.name="Students"
print(s1)
```

**Attribute**:value  
**Description**:Returns the list of values in the series

```python
print(s1.values)
```

**Attribute**:size  
**Description**:Returns the number of values in the series

```python
print(s1.size)
```

**Attribute**:empty
**Description**:Returns True if the series is empty

```python
print(s1.empty)
```
