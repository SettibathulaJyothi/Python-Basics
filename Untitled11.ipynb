**Type Conversion**  
Is the process of converting value of one data type to another

```python
a=5
print("The datatype of a is:",type(a))
b=0.536576588142458
print("The datatype of a is:",type(b))
c=5.5954657658965865386896575623748565575164964463
print("The datatype of a is:",type(c))
```
